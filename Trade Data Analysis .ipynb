```python
df["PnL"] = df["Port_IDs_Encode"]  

# First 5 rows check karo  
print(df[["Port_IDs_Encode", "PnL"]].head())
```

```python
df["PnL"] = df["Trade_History_Encode"]  

#7. First 5 rows check karo  
print(df[["Trade_History_Encode", "PnL"]].head())
```

```python
  
#8. Win ya Loss identify   
df["Win"] = df["PnL"].apply(lambda x: 1 if x > 0 else 0)  
df["Loss"] = df["PnL"].apply(lambda x: 1 if x < 0 else 0)  

#9. Total positions count   
total_positions = len(df)  

#10. Win positions count 
win_positions = df["Win"].sum()  

#11. Win Rate calculate   
win_rate = (win_positions / total_positions) * 100 if total_positions > 0 else 0  

#12. Print results  
print(f"Total Positions: {total_positions}")
print(f"Win Positions: {win_positions}")
print(f"Win Rate: {win_rate:.2f}%")
```

```python
#13. Cumulative PnL calculate 
df["Cumulative PnL"] = df["PnL"].cumsum()  

#14. Maximum value track 
df["Cumulative Max"] = df["Cumulative PnL"].cummax()  

#15. Drawdown calculate 
df["Drawdown"] = df["Cumulative PnL"] - df["Cumulative Max"]  

#16. MDD ka minimum value  
mdd = df["Drawdown"].min()  

print(f"Maximum Drawdown (MDD): {mdd:.2f}")
```

```python
#17. Daily Returns calculate  
daily_returns = df["PnL"].pct_change().dropna()  

#18. Sharpe Ratio calculate  
sharpe_ratio = daily_returns.mean() / daily_returns.std() if daily_returns.std() != 0 else 0  

# 19. Annualize  
sharpe_ratio *= np.sqrt(252)  

print(f"Sharpe Ratio: {sharpe_ratio:.2f}")

```

```python
#20. Final CSV file save 
df.to_csv("trade_data_with_metrics.csv", index=False)  

#21. Summary Metrics Save
metrics_df = pd.DataFrame({
    "Total Positions": [total_positions],
    "Win Positions": [win_positions],
    "Win Rate (%)": [win_rate],
    "Max Drawdown (MDD)": [mdd],
    "Sharpe Ratio": [sharpe_ratio],
           })
metrics_df.to_csv("financial_metrics.csv", index=False)

print(" Metrics calculation complete. Check trade_data_with_metrics.csv & financial_metrics.csv")
```
